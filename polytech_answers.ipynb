## Import libraries below

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

## Read file

```python
df = pd.read_csv('raw_data.csv')
```

## Inspect all data

```python
df.info()
```

## Rename column name

```python
df.rename(
    columns={
        'Treatment': 'treatment',
        'COMPANY SIZE': 'company_size',
        'Mental health support': 'mental_health_support'
    }, inplace=True
)

df.info()
```

```python
df.columns
```

## Change all missing numbers to "Others"

```python
df['gender'] = df['gender'].fillna("Others")
df.info()
```

## Add a column named "dummy_column"

```python
df['dummy_column'] = np.nan
df
```

## Drop a column

```python
df.drop(columns=['dummy_column'], inplace=True)
df
```

## Export final data (cleaned)

```python
df.to_csv("cleaned_data.csv", index=False) 
df = pd.read_csv("cleaned_data.csv")  
df
```

## Count the number of people who received treatment, grouped by gender

```python
treatment_count = df[df['treatment'] == 'Yes']
gender_counts = treatment_count['gender'].value_counts()
gender_counts
```

```python
treatment_count = df[df['treatment'] == 'Yes']

plt.figure(figsize=(10, 4))
gender_counts = treatment_count['gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values, color=['pink', 'blue', 'red'])
plt.title('Number of individuals seeking mental health treatment by gender')
plt.xlabel('Gender')
plt.ylabel('Number of genders of who received treatment')
plt.legend()
plt.show()
```

# Insights based on final visualization

- More women are seeking mental health support, suggesting they may be more open to asking for help.
- Fewer men are doing so, possibly due to stigma or hesitation.
- Those in the others category also seek help, highlighting the need for inclusive, accessible support for all.
